The dataframe will contain date from 2 october 2017 to 31 march 2019

```python
import pandas as pd
import numpy as np

start_date = '2017-09-02'
end_date = '2019-03-31'
mydates = pd.date_range(start_date, end_date).tolist()
df = pd.DataFrame(np.array(mydates).reshape(576,1), columns = ["Dates"])
```
